# Generating the training and testing sets

```python
df_final['Attrition'].value_counts()
```

#### Thus it is an imbalanced dataset
We need to fix it

```python
X = df_final.drop('Attrition', axis=1)
y = df_final['Attrition']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train, X_test
```

# Testing models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

## Logistic Regression

```python
lr = LogisticRegression(solver='liblinear')
params = {'C':[1e-5, 1e-3, 0.01, 0.1, 1, 10, 50, 100], 'penalty':['l1','l2']}
clf = GridSearchCV(lr, params, cv=5, scoring='f1_macro')
```

#### If class imbalance is the concern use f1_macro

```python
clf.fit(X_train, y_train)
```

```python
print(clf.best_params_)
print(clf.best_score_)
```

#### f1_macro score is good

```python
y_pred = clf.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

####
1. Accuracy is very good but the dataset is imbalanced
2. False Positive (FP) = 0
3. False Negative (FN) is high. Need to reduce it
4. True Positive (TP) is very low = 1. Need to increase it
5. Recall and f1_score wrt 1 is very poor

### Using class_weight on Logistic Regression

```python
lr = LogisticRegression(class_weight='balanced', solver='liblinear')
params = {'C':[1e-5, 1e-3, 0.01, 0.1, 1, 10, 50, 100], 'penalty':['l1','l2']}
clf = GridSearchCV(lr, params, cv=5, scoring='f1_macro')
```

```python
clf.fit(X_train, y_train)
```

```python
print(clf.best_params_)
print(clf.best_score_)
```

#### f1_macro score has increased but it is still bad.

```python
y_pred = clf.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### Tuning class_weight hyperparameter

```python
lr = LogisticRegression(solver='liblinear')
weights = np.linspace(0.05, 0.95, 100)
params = {'C':[1e-5, 1e-3, 0.01, 0.1, 1, 10, 50, 100], 'class_weight':[{0: x, 1: 1.0-x} for x in weights], 'penalty':['l1','l2']}
clf = GridSearchCV(lr, params, cv=5, scoring='f1_macro', n_jobs=-1)
```

```python
clf.fit(X_train, y_train)
```

```python
print(clf.best_params_)
print(clf.best_score_)
```

```python
y_pred = clf.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

#### Good balance between precision and recall

### Sampling to reduce the imbalance in the dataset

```python
lr = LogisticRegression(solver='liblinear')
params = {'C':[1e-5, 1e-3, 0.01, 0.1, 1, 10, 50, 100], 'penalty':['l1','l2']}
clf = GridSearchCV(lr, params, cv=5, scoring='f1_macro', n_jobs=-1)
```

```python
clf.fit(X_train_ns, y_train_ns)
```

```python
print(clf.best_params_)
print(clf.best_score_)
```

```python
y_pred = clf.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

####
1. Very poor accuracy, precision and f1_score wrt 1
2. FP has drastically increased
3. Undersampling is not a good method

#### Over Sampling

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
os = RandomOverSampler(sampling_strategy=0.5)              # 853 * 0.5 = 426
X_train_os, y_train_os = os.fit_resample(X_train, y_train)
print("The number of classes before the fit {}".format(Counter(y_train)))
print("The number of classes after the fit {}".format(Counter(y_train_os)))
```

```python
lr = LogisticRegression(solver='liblinear')
params = {'C':[1e-5, 1e-3, 0.01, 0.1, 1, 10, 50, 100], 'penalty':['l1','l2']}
clf = GridSearchCV(lr, params, cv=5, n_jobs=-1, scoring='f1_macro')
```

```python
clf.fit(X_train_os, y_train_os)
```

```python
print(clf.best_params_)
print(clf.best_score_)
```

```python
y_pred = clf.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

#### Comparing to undersampling

### Hyperparameter tuning and observing error-rate graph for optimal 'k' value

```python
err = []
for i in range(1,50):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    i_pred = knn.predict(X_test)
    err.append(np.mean(i_pred != y_test))
```

```python
plt.figure(figsize=(15,10))
plt.plot(range(1, 50), err)
```

#### This shows that the optimal k = 10

### Using k=10 and no other hyperparameter tuning and no sampling
